## Finding the extrema of a function

```python
tStart = 0
tEnd = 10
tNumber = 100
tEval = np.linspace(tStart, tEnd, tNumber)
maxStep = (tEnd - tStart) / tNumber

def fprime2(t, y):
    deriv = [0, 0]
    deriv[0] = np.cos(t)
    deriv[1] = -np.sin(t)
    
    return deriv

def disZeroEvent(t, y):
    return y[0]

def firstDerZeroEvent(t, y):
    return y[1]

sol2 = solve_ivp(fprime2, (tStart, tEnd), [0, 1], t_eval=tEval, max_step=maxStep, events=[disZeroEvent, firstDerZeroEvent])

# print('y events: ' + str(sol2.y_events))
# print('y events[0]: ' + str(sol2.y_events[0]))
# print('y events[0][0]: ' + str(sol2.y_events[0][0]))
# print('y events[0][0][0]: ' + str(sol2.y_events[0][0][0]))
# print('t events: ' + str(sol2.t_events))

# graph of displacement vs time
plt.plot(sol2.t, sol2.y[0])
plt.plot(sol2.t_events[0], sol2.y_events[0][:, 0], 'o')
plt.plot(sol2.t_events[1], sol2.y_events[1][:, 0], 'x')
plt.legend(('function', 'zeros', 'extrema'))
plt.ylabel('displacement')
plt.xlabel('time')
plt.show()
```

## Finding the inflection points of a function

```python
tStart = -5
tEnd = 0.5
tNumber = 100
tEval = np.linspace(tStart, tEnd, tNumber)
maxStep = (tEnd - tStart) / tNumber

def fprime3(t, y):
    deriv = [0, 0, 0]
    deriv[0] = 1.5 * t**2 + 7*t + 28/5
    deriv[1] = 3*t + 7
    deriv[2] = 3
    
    return deriv

def zeroEvent(t, y):
    return y[0]

def extremaEvent(t, y):
    return y[1]

def inflectionEvent(t, y):
    return y[2]

sol3 = solve_ivp(fprime3, (tStart, tEnd), [0, 8.1, -8], t_eval=tEval, max_step=maxStep, events=[zeroEvent, extremaEvent, inflectionEvent])

plt.plot(sol3.t, sol3.y[0])
plt.plot(sol3.t_events[0], sol3.y_events[0][:, 0], 'o')
plt.plot(sol3.t_events[1], sol3.y_events[1][:, 0], 'o')
plt.plot(sol3.t_events[2], sol3.y_events[2][:, 0], 'o')
plt.legend(('function', 'zeros', 'extrema', 'inflection'))
plt.show()
```
