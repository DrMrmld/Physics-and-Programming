# Stats about COVID-19 victims
## Inserting the database

The code below imports the .csv database with all the data about COVID-19 victims. `dfCovid.drop()` function removes certain columns from the original database.

Source: [Our World in Data: COVID-19 data (GitHub repository)](https://github.com/owid/covid-19-data/tree/master/public/data)

```python
import numpy as np
import pandas as pd

from sklearn.impute import SimpleImputer as simp

dfCovid = pd.read_csv('owid-covid-data.csv')

dfCovid.drop(['iso_code', 'continent', 'new_cases_smoothed', 'new_cases_smoothed_per_million', 'new_deaths_smoothed', 'new_deaths_smoothed_per_million'], axis=1, inplace=True)
dfCovid
```

## Data grouping

The code below groups all rows that have similar `location` value by summing all the values and reseting indices from 0. The data is not correct (Africa does not have 1.7 bln cases of COVID-19) because of the way that the database was structured — for every `total_cases` (and other) value they've written a cumulative value already.

```python
dfCovidByCountryCumulative = dfCovid.groupby('location')[['total_cases', 'total_cases_per_million', 'total_deaths', 'total_deaths_per_million']].sum().reset_index()
dfCovidByCountryCumulative.head(20)
```

## Data Plots
### Total cases by country

New `dfCovidByCountry` data frame takes all the rows that have a `date` value of `2021-09-25` and the indicated fields.

`dfCovidForCountryName` data frame takes all the rows that have a `location` value of `United States` (and all the fields from the original data frame). `len(dfCovidForCountryName['date'])` gives the total number of days after the start of the pandemic. The program plots a scatter of days after the start of the pandemic versus the number of total cases (and ~deaths) for the selected country.

```python
import matplotlib.pyplot as plt

dfCovidByCountry = dfCovid[dfCovid['date'] == '2021-09-25'].reset_index()
dfCovidByCountry2 = dfCovidByCountry[['location', 'date', 'total_cases', 'total_deaths', 'icu_patients', 'people_vaccinated', 'median_age', 'aged_65_older']]
dfCovidByCountry2[150:200] #shows total number of cases, deaths, etc. for 2021-09-25

countries = dfCovid['location'].unique()
print('Number of countries in dfCovid data frame:', len(countries))
#print(countries)

countries2 = dfCovidByCountry['location'].unique()
print('Number of countries in dfCovidByCountry data frame:', len(countries2))
#print(countries2)

print('Some countries are lost because they do not have any data for the date 2021-09-25')



# error = []
# for i in countries:
#     for j in countries2:
#         if i == j:
#             break
#     else:
#         error.append(i)

# print(error) --- prints the countries that are not in first list

# dfCovid[(dfCovid['location'] == 'Anguilla')] --- one of the 'lost' countries

countryName = 'United States'
dfCovidForCountryName = dfCovid[dfCovid['location'] == countryName].reset_index()
plt.scatter(np.arange(len(dfCovidForCountryName['date'])), dfCovidForCountryName['total_cases'], s=5, label='Total cases')
plt.scatter(np.arange(len(dfCovidForCountryName['date'])), dfCovidForCountryName['total_deaths'] * 10, s=5, label='Total deaths * 10')
plt.title(countryName)
plt.xlabel('Days from 2/24/2020')
plt.ylabel('Total number')
plt.legend()
plt.show()

countryName = 'Russia'
dfCovidForCountryName = dfCovid[dfCovid['location'] == countryName].reset_index()
plt.scatter(np.arange(len(dfCovidForCountryName['date'])), dfCovidForCountryName['total_cases'], s=5, label='Total cases')
plt.scatter(np.arange(len(dfCovidForCountryName['date'])), dfCovidForCountryName['total_deaths'] * 10, s=5, label='Total deaths * 10')
plt.title(countryName)
plt.xlabel('Days from 2/24/2020')
plt.ylabel('Total number')
plt.legend()
plt.show()

countryName = 'Kazakhstan'
dfCovidForCountryName = dfCovid[dfCovid['location'] == countryName].reset_index()
plt.scatter(np.arange(len(dfCovidForCountryName['date'])), dfCovidForCountryName['total_cases'], s=5, label='Total cases')
plt.scatter(np.arange(len(dfCovidForCountryName['date'])), dfCovidForCountryName['total_deaths'] * 10, s=5, label='Total deaths * 10')
plt.title(countryName)
plt.xlabel('Days from 2/24/2020')
plt.ylabel('Total number')
plt.legend()
plt.show()

countryName = 'Brazil'
dfCovidForCountryName = dfCovid[dfCovid['location'] == countryName].reset_index()
plt.scatter(np.arange(len(dfCovidForCountryName['date'])), dfCovidForCountryName['total_cases'], s=5, label='Total cases')
plt.scatter(np.arange(len(dfCovidForCountryName['date'])), dfCovidForCountryName['total_deaths'] * 10, s=5, label='Total deaths * 10')
plt.title(countryName)
plt.xlabel('Days from 2/24/2020')
plt.ylabel('Total number')
plt.legend()
plt.show()
```

### New cases by country

This code does the same thing as the code above. The difference is that this code plots the new cases for a set of countries, and it is also simpler because of the use of the loop.

```python
countriesSet = ['Brazil', 'United States', 'Kazakhstan', 'Russia']

for x in countriesSet:
    countryName = x
    dfCovidForCountryName = dfCovid[dfCovid['location'] == countryName].reset_index()
    plt.scatter(dfCovidForCountryName['date'], dfCovidForCountryName['new_cases'], s=5)
    plt.title('Number of new cases per day for ' + x)
    plt.xlabel('Days from 2/24/2020')
    plt.ylabel('Number of new cases per day')
    plt.show()
```

### New cases by country comparison

```python
countriesSet = ['Brazil', 'United States', 'Kazakhstan', 'Russia']

for x in countriesSet:
    countryName = x
    dfCovidForCountryName = dfCovid[dfCovid['location'] == countryName].reset_index()
    plt.scatter(dfCovidForCountryName['date'], dfCovidForCountryName['new_cases'], s=5, label=x)
    
plt.title('Number of new cases per day for different countries')
plt.xlabel('Days from 2/24/2020')
plt.ylabel('Number of new cases per day')
plt.ylim([-10000, 320000])
plt.legend()
plt.show()
```

### New cases per million by country comparison

```python
countriesSet = ['Brazil', 'United States', 'Kazakhstan', 'Russia']

for x in countriesSet:
    countryName = x
    dfCovidForCountryName = dfCovid[dfCovid['location'] == countryName].reset_index()    
    plt.scatter(dfCovidForCountryName['date'], dfCovidForCountryName['new_cases_per_million'], s=5, label=x)
    
plt.title('Number of new cases per day per million for different countries')
plt.xlabel('Days from 2/24/2020')
plt.ylabel('Number of new cases per day per million')
plt.ylim([-20, 1000]) #gets rid of the outliers in Kazakhstan
plt.legend()
plt.show()
```

### GDP per capita versus total cases

```python
dfCovidGDP = dfCovid[dfCovid['date'] == '2021-09-25'].reset_index()
dfCovidGDP = dfCovidGDP[['location', 'total_cases_per_million', 'total_deaths_per_million', 'gdp_per_capita']]
dfCovidGDP_modified = dfCovidGDP[dfCovidGDP['gdp_per_capita'] < 20000].reset_index()
#print(dfCovidGDP.head(20))

plt.scatter(dfCovidGDP['gdp_per_capita'], dfCovidGDP['total_cases_per_million'], s=5)
plt.title('GDP per capita vs total cases per million')
plt.xlabel('GDP per capita')
plt.ylabel('Total cases per million')
```

```python
plt.scatter(dfCovidGDP_modified['gdp_per_capita'], dfCovidGDP_modified['total_cases_per_million'], s=5, color='orange')
plt.title('GDP per capita vs total cases per million (GDP<20000)')
plt.xlabel('GDP per capita')
plt.ylabel('Total cases per million')
```
